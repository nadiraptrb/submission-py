## Exploratory Data Analysis (EDA)

### Explore ...

```python
print(day_df.columns)
```

```python
data_2011_2012 = day_df[(day_df["today's date"].dt.year >= 2011) & (day_df["today's date"].dt.year <= 2012)]
```

```python
day_df.describe(include="all")
```

```python
print(day_df['season'].unique())
```

```python
# Membuat kolom baru untuk musim berdasarkan bulan
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

data_2011_2012['season'] = data_2011_2012["today's date"].dt.month.apply(get_season)
```

```python
# Analisis Korelasi antara Musim dan Jumlah Penyewaan Sepeda
# Menghitung korelasi antara musim dan jumlah penyewaan sepeda
season_counts = data_2011_2012.groupby('season')['cnt'].mean()
```

```python
hourly_usage = hour_df.groupby('hr')['cnt'].mean()
```

```python
hour_df.describe()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

Visualisasi hasil analisis batang bar plot


```python
plt.figure(figsize=(10, 6))
sns.barplot(x=season_counts.index, y=season_counts.values, color='darkblue')
plt.title('Average Bike Usage by Season (2011-2012)')
plt.xlabel('Season')
plt.ylabel('Average Bike Usage')
plt.show()
```

```python
# Memfilter data untuk musim spring
spring_data = data_2011_2012[data_2011_2012['season'] == 'Spring']

# Menghitung jumlah pengguna sepeda di musim spring
spring_bike_usage = spring_data['cnt'].sum()

# Memfilter data untuk musim summer
summer_data = data_2011_2012[data_2011_2012['season'] == 'Summer']

# Menghitung jumlah pengguna sepeda di musim summer
summer_bike_usage = summer_data['cnt'].sum()

# Memfilter data untuk musim fall
fall_data = data_2011_2012[data_2011_2012['season'] == 'Fall']

# Menghitung jumlah pengguna sepeda di musim fall
fall_bike_usage = fall_data['cnt'].sum()

# Memfilter data untuk musim winter
winter_data = data_2011_2012[data_2011_2012['season'] == 'Winter']

# Menghitung jumlah pengguna sepeda di musim winter
winter_bike_usage = winter_data['cnt'].sum()

# Tampilkan jumlah pengguna sepeda di setiap musim
print("Spring Bike Usage:", spring_bike_usage)
print("Summer Bike Usage:", summer_bike_usage)
print("Fall Bike Usage:", fall_bike_usage)
print("Winter Bike Usage:", winter_bike_usage)
```

### Pertanyaan 2:

Visualisasi hasil analisis dengan line plot

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=hourly_usage.index, y=hourly_usage.values, marker='o')
plt.title('Average Bike Usage by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Bike Usage')
plt.grid(True)
plt.show()
```

```python
# Analisis Perbedaan Antara Jam Sibuk dan Jam Sepi
# Mengelompokkan data berdasarkan jam operasional harian (hour) dan menghitung rata-rata jumlah penyewaan sepeda (cnt)
# Kemudian membagi menjadi jam sibuk (misalnya, jam 7-10 pagi dan jam 4-7 sore) dan jam sepi (sisanya)
busy_hours = hour_df[(hour_df['hr'].isin(range(7, 11)) | hour_df['hr'].isin(range(16, 20)))].groupby('hr')['cnt'].mean().mean()
non_busy_hours = hour_df[~((hour_df['hr'].isin(range(7, 11))) | (hour_df['hr'].isin(range(16, 20))))]['cnt'].mean()
```

```python
# Tampilkan perbedaan tingkat penggunaan sepeda antara jam sibuk dan jam sepi
print("Rata-rata Penggunaan Sepeda pada Jam Sibuk:", busy_hours)
print("Rata-rata Penggunaan Sepeda pada Jam Sepi:", non_busy_hours)
```
